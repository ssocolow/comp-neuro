```python
import matplotlib.pyplot as plt
```

```python
def plotAllTopClasses(d):
    fig, axes = plt.subplots(16, 16, figsize=(20, 20))
    fig.suptitle('Distribution of Top 100 Activated Classes for Each Neuron', fontsize=16)

    for i, ax in enumerate(axes.flatten()):
        if i < len(d):
            neuron_classes = d[i]
            ax.hist(neuron_classes, bins=range(11), align='left', rwidth=0.8)
            ax.set_title(f'Neuron {i}')
            ax.set_xticks(range(10))
            ax.tick_params(axis='x', labelsize=8)
            ax.tick_params(axis='y', labelsize=8)
        else:
            ax.axis('off') # Turn off extra subplots if there are fewer than 256 neurons

    plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent title overlap
    plt.show()
```

```python
rs = [[1 for i in range(100)] for i in range(256)]
```

```python
plotAllTopClasses(rs)
```
